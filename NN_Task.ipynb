# Load the Dataset

```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load the dataset
url = 'diabetes.csv'
df = pd.read_csv(url)
df.head()
```

# Preprocess the Data

```python
#Handle Missing or Inconsistent Data

# Replace zeros with NaN for specific columns where 0 is not a valid value
df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0, pd.NA)

# Fill missing values with the median (you can also use mean or other methods)
df.fillna(df.median(), inplace=True)
```

```python
#Standardize the Feature Values

# Separating features and target variable
X = df.drop(columns=['Outcome'])
y = df['Outcome']

# Standardizing the feature values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Build a Neural Network Model

```python
# Building the neural network model
model = Sequential()

# Input layer (8 features) and one hidden layer
model.add(Dense(16, input_shape=(X_scaled.shape[1],), activation='relu'))

# Adding another hidden layer
model.add(Dense(8, activation='relu'))

# Output layer (for binary classification)
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Summary of the model
model.summary()
```

# Train the Model

```python
# Split the data into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test))
```

# Evaluate the Model

```python
# Make predictions on the test data
y_pred = (model.predict(X_test) > 0.5).astype("int32")

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

# Hyperparameter Tuning


```python
# Example of changing hyperparameters
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model with new hyperparameters
history_after_tuning = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Predict on the test data after tuning
y_pred_after_tuning = (model.predict(X_test) > 0.5).astype("int32")

# Import necessary libraries for performance metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculate accuracy after tuning
accuracy_after_tuning = accuracy_score(y_test, y_pred_after_tuning)
print(f"Accuracy after tuning: {accuracy_after_tuning:.2f}")

# Calculate precision, recall, and F1-score after tuning
precision_after_tuning = precision_score(y_test, y_pred_after_tuning)
recall_after_tuning = recall_score(y_test, y_pred_after_tuning)
f1_after_tuning = f1_score(y_test, y_pred_after_tuning)

print(f"Precision after tuning: {precision_after_tuning:.2f}")
print(f"Recall after tuning: {recall_after_tuning:.2f}")
print(f"F1 score after tuning: {f1_after_tuning:.2f}")
```
